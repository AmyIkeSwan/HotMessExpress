```python
## scatterplot of Introversion and Extroversion by cohort
### using pandas
Plot=f2fmerged.plot.scatter(x='Sum_Introversion', y='Sum_Extroversion', c='Cohort')
Plot.set_title('Self-Ratings for Introversion and Extroversion by Cohort')
Plot.set_xlabel('Introversion')
Plot.set_ylabel('Extroversion')
```

```python
## scatterplot of Introversion and Extroversion by cohort
### using pyplot
groups = f2fmerged.groupby('Cohort')
for name, group in groups:
    plt.plot(group.Sum_Introversion, group.Sum_Extroversion, marker='o', linestyle='', markersize=3, label=name)
    plt.legend()
```

```python
## scatterplot of Introversion and Extroversion by cohort
### using pyplot
plt.figure(figsize=(8,8))
sp_names = ['2019', '2020', '2021']
scatter = plt.scatter(f2fmerged.Sum_Introversion, 
                      f2fmerged.Sum_Extroversion, 
                      s=150,
                      c=f2fmerged.Cohort.astype('category').cat.codes)
plt.xlabel("Introversion Score", size=14)
plt.ylabel("Extroversion Score", size=14)
plt.legend(handles=scatter.legend_elements()[0],
           labels=sp_names, 
           title="Cohort")
plt.savefig("Self-Rated Introversion and Extroversion Scores by Cohort.png",
          format='png', dpi=150)
```

```python
## scatterplot of Introversion and Extroversion by cohort
### using seaborn
sns.relplot(
data=f2fmerged, x="Sum_Introversion", y="Sum_Extroversion",
col="Cohort", style="Cohort",
kind="scatter")
```
